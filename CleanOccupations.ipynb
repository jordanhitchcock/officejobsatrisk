Clean and transform series data to be used in plotting and analysis. Reduce size to <100MB so that it meets Github's file size limits.

```python
data = pd.read_csv('oe.data.0.Current.txt', sep='\t', dtype='string')
data.columns = data.columns.str.strip()
data['value'] = data['value'].str.replace('\s*\-', '0', regex=True)  # replace values consisting of whitespace and a dash with a 0
data['value'] = pd.to_numeric(data['value'])

data['series_id'] = data['series_id'].str.strip()
data['seasonal'] = data['series_id'].str.slice(2, 3)
data['areatype_code'] = data['series_id'].str.slice(3, 4)
data['area_code'] = data['series_id'].str.slice(4, 11)
data['industry_code'] = data['series_id'].str.slice(11, 17)
data['occupation_code'] = data['series_id'].str.slice(17, 23)
data['datatype_code'] = data['series_id'].str.slice(23)

office_data = data[data['occupation_code'].isin(office_occupations['occupation_code'])]

metro_employment = office_data.loc[(data['areatype_code'] == 'M') &
                                   (data['datatype_code'] == '01') &
                                   (data['seasonal'] == 'U'),
                                  ['area_code', 'industry_code', 'occupation_code', 'value']]
```

```python
metro_employment.to_csv('metro_employment.csv', index=False)
```
